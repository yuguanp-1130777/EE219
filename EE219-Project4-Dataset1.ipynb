### For some other cases, e.g. when encoding time stamps such as {Mon, ..., Sun} or {Jan, ..., Dec} it might make sense to perform either one. In those cases the learning algorithm of choice and numerical results can lead our way. Can you explain a trade-off here? (Hint: let us assume we perform linear regression, what information does one-hot encoding discard, and what assumption should hold strongly if we perform the scalar encoding instead?) (Question 7)

If we use a one-hot encoding instead of scalar, then we disgard the numerical relationship. For example, if poor is 0, good is 1 and outstanding is 2. If we use one-hot encoding, then we disgard the information that outstanding is better (greater) than good and good is greater (higher) than bad that scaler can convey. Therefore, if we use scalar enconding, we must guarantee that the order/degree of the categories can be represented by the numerical values. If there is no such order exist, then we should use one-hot encoding.

For this bike-share dataset, we will one-hot encoding season, month, weekday and keep the scalar encoding of weather situation.

### Explain how each regularization scheme affects the learned hypotheses. (Question 10)

After adding regularization terms, we are minimizing the sum of least square error and regularization terms. L1-regularization use L1 norm for each parameters and L2-regularization use L2 norms. With regularization term, the model will force parameters of some features to be zero (in L1 regularization) or close to zero (in L2 regularization). Feature with small absolute value of parameters will have limited impact on model. Regularization can keep model simple to avoid overfitting.

### Does feature scaling play any role (in the cases with and without regularization)? Justify your answer. (Question 12)

When using OLR model, the feature scaling does not play much role. However, with regularization, the feature scaling is important, because the range of features can siginificantly affect the magnitude of parameters. If parameters are on different magnitude, it will be difficult for regularization to optimize since parameters with large values will be more attrative than parameters with small values. In that case, we are mostly just regularizing parameters with large values instead of fairly evaluating all the features.

### Some linear regression packages return p-values for different features. What is the meaning of them and how can you infer the most significant features? (Question 13)

The p-values means the probability of the feature is not significant different from 0. After calculating the parameters, we want to evaluate the normailzed distance between the parameter with 0. If the value is too close to 0, that means that feature may not be important to the model since the probability of it's 0 is high. p-values shows the probability of the parameter should be 0 based on the normalized distance of parameter value away from 0, and a high p value (for example, p value > 0.1) means the feature is not important when prediction with a high chance. We can regard features with P value < 0.05 (or more a strict threshold) as significant features.

### For random forest model, measure Out-of-Bag Error (OOB) as well. Explain what OOB error and R2 score means given this link. (Question 28)

```python
oob_error = 1 - best_rf.oob_score_
print("The OOS Error for RF model is: " + str(oob_error))
```

R2 means the percentage of information the model learned compared to a naive model. A naive model means no matter what's the input is, the model will only output average of training labels. We use the squared error of this model compared to grouth truth as a benchmark and regard the squared error of this model as all the information a model should learn. R-squared measure the percentage of information leanred by a given model. If R-squared is 99%, that means the model learns 99% of the information which naive model does not learn. R-squared can only between 0 and 1 and the closer to 1 the better.

OOB error is a measure for bagging methods. When bootstrap aggregating is performed, we will have the bootstrap sample - namely the data chosen to be "in-the-bag" by sampling with replacement. The out-of-bag set is all data not chosen in the sampling process. After train the model on bootstrap set, we can test the model on OOB set to evaluate the model. Since the tree is not trained on OOB set, it is a good test for the performance of the model. After we finish predicting OOB samples with the tree not trained on them, take the majority vote of these models' result for the OOB instance and compare to the true value of the OOB instance to calculate the OOB error for all instances in the OOB dataset. Once the OOB error stabilizes, it will converge to cross-validation error.
